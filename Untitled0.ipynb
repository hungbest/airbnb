```python
import math
# Ham dung chung
def evenFunction(x, f):
  if(f(-x)==f(x)):
    return True
  else:
    return False

def oddFunction(x, f):
  if(f(-x) == -f(x)):
    return True
  else:
    return False

# is Increasing Function
def increasingFunction(x1, x2, f):
  if(x1 < x2):
    if(f(x1)< f(x2)):
      return True
    else:
      return False
  return False

# is Descreasing Function
def decreasingFunction(x1, x2, f):
  if(x1 < x2):
    if(f(x1) > f(x2)):
      return True
    else:
      return False
  return False


# EXERCISE 4
from sympy import *
x0 = symbols('x')
root = solve(x0**2 + 2*x0 -1)
print("root: {}".format(root))
print("root: {}, {}".format(root[0].evalf(), root[1].evalf())) #convert from symbol to float 


## EXERCISE 3
# a
print('Bai a:')
x = 3
fa = lambda x: x**-5
if(evenFunction(x, fa)==True):
  print('Ham chan')
if(oddFunction(x, fa)==True):
  print('Ham le')

# b
print('Bai b:')
x = 8
fb = lambda x: x**2 + x
if(evenFunction(x, fb)==True):
  print('Ham chan')
if(oddFunction(x, fb)==True):
  print('Ham le')

# c
print('Bai c:')
x = 2
fc = lambda x: x**2 + 1
if(evenFunction(x, fc)==True):
  print('Ham chan')
if(oddFunction(x, fc)==True):
  print('Ham le')

# d
print('Bai d:')
x = 5
fd = lambda x: x**3 + x
if(evenFunction(x, fd)==true):
  print('Ham chan')
if(oddFunction(x, fd)==True):
  print('Ham le')

# e
print('Bai e:')
x = 4
fe = lambda x: abs(math.pow(x, 3))
if(evenFunction(x, fe)==True):
  print('Ham chan')
if(oddFunction(x, fe)==True):
  print('Ham le')

# f
print('Bai f:')
x = 2
ff = lambda x: 2*x + 1
if(evenFunction(x, ff)==True):
  print('Ham chan')
if(oddFunction(x, ff)==True):
  print('Ham le')

# g
print('Bai g:')
x = 2
fg = lambda x: math.pow(x, 4) + 3*x**2 - 1
if(evenFunction(x, fg)==True):
  print('Ham chan')
if(oddFunction(x, fg)==True):
  print('Ham le')

# h
print('Bai h:')
x = 5
fh = lambda x: 2*abs(x) + 1
if(evenFunction(x, fh)==True):
  print('Ham chan')
if(oddFunction(x, fh)==True):
  print('Ham le')


# --------------------------------------------------
# if(increasingFunction(x1, x2, fa) == True):
#   print('Ham tang')
# if(decreasingFunction(x1, x2, fa) == True):
#   print('Ham giam')
```

```python
from sympy import *
x0 = symbols('x')
f5v = sqrt(abs(x0))
plot(f5v, (x0, -4, 4), line_color = "green", title = "sqrt(abs(x0))", show = true)

print('==================================================')

x1 = symbols('x')
f5b = pow(x1, 4) + 3*pow(x1, 2) - 1
plot(f5b, (x1, -5, 5), line_color = "cyan", title = "pow(x1, 4) + 3*pow(x1, 2) - 1", show = true)

print('==================================================')

x2 = symbols('x')
f5c = pow(x2, 3) + x2
plot(f5c, (x2, -5, 5), line_color = "yellow", title = "pow(x2, 3) + x2", show = true)

print('==================================================')

x3 = symbols('x')
f5d = exp(x3)
plot(f5d, (x3, -5, 5), line_color = "red", title = "exp(x3)", show = true)

print('==================================================')

x4 = symbols('x')
f5e = log(x4)
plot(f5e, (x4, -5, 5), line_color = "orange", title = "log(x4)", show = true)

print('==================================================')

x5 = symbols('x')
f5f = (2*x5**2 -3)/(7*x5 + 4)
plot(f5f, (x5, -5, 5), line_color = "blue", title = "(2*x5**2 -3)/(7*x5 + 4)", show = true)


```
